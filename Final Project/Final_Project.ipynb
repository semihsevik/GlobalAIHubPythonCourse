```python
import pandas as pd
import time
```

```python
def createQandA():
    Q0,A0= "Who invented the Turing Test ?: " , "Alan Turing"
    Q1,A1 = "Derivative of sin(x) ?: " , "cos(x)"
    Q2,A2 = "Who invented the Fuzzy Logic ?: " , "L端tfi Askerzade"
    Q3,A3 = "Derivative of x^2 ?: " , "2x"
    Q4,A4 = "Which city in the world has the highest population ?: " , "Tokyo"
    Q5,A5 = "Who won the first FIFA World Cup ?: " , "Uruguay"
    Q6,A6 = "Python designed by Guido van Rossum.(T/F): " , "T"
    Q7,A7 = "Who is the world chess champion (currently) ?: " , "Magnus Carlsen"
    Q8,A8 = "Scikit-learn is a Python library used for Machine Learning. (T/F): " , "T"
    Q9,A9 = "C is an object oriented programming language. (T/F): " , "F"

    Qlist = [Q0,Q1,Q2,Q3,Q4,Q5,Q6,Q7,Q8,Q9]
    Alist = [A0,A1,A2,A3,A4,A5,A6,A7,A8,A9]

    return Qlist , Alist
```

```python
Qlist , Alist = createQandA()
myDict = {"Q" : Qlist , "A" : Alist}
df = pd.DataFrame.from_dict(myDict) #For better visualization
df
```

```python
competitors = []
result = " "
print("\n******* Welcome to our Knowledge Competition ! (Press 'q' to exit.) *******\n")

while True:
    score = 0
    username = input("Name: ")
    
    if username != 'q':
        
        for i in range(len(myDict['Q'])):
            Q = myDict['Q'][i]
            A = myDict['A'][i]
            userAnswer = input(f"\nQ-{i+1}: {Q}")

            if userAnswer == A or userAnswer == A.lower():
                score += 10
                print(f"Congrats ! Score --> {score}")
            
            else: print(f"Wrong Answer ! Score --> {score}")
        
        print(f"Congrats {username} ! Your total score --> {score}\n")

        if score > 50:
            result = "Succesful"
        else: result = "Unsuccesful"

        tup = (score,username,result)
        competitors.append(tup)

    else: 
        print("Saving results...")
        time.sleep(2)
        competitors.sort(reverse=True)
        print("Saved !")
        break
        

print("\n******* Our Knowledge Competition has ended ! ******* \n")
for i in competitors:
    print(f"{i[1]} --> Score: {i[0]} --> Result: {i[2]}\n")

champName = competitors[0][1]
champScore = competitors[0][0]
print(f"******************\n Champion: {champName} !\n Score: {champScore}\n******************")
```
